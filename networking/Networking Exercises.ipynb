# Getting your own IP address

```python
import socket
host = socket.gethostname()
ip = socket.gethostbyname(host)
print(host, ip)
```

# Getting the IP address of a domain

```python
try:
    print(socket.gethostbyname('www.google.com'))
except socket.error as err:
    print(err)
```

# Getting all IP addresses associated with all network interfaces

```python
import psutil

addrs = psutil.net_if_addrs()
ip_addresses = []

for interface, addr_list in addrs.items():
    for addr in addr_list:
        if addr.family == socket.AF_INET:
            ip_addresses.append((interface, addr.address))

print("IP Addresses:", ip_addresses)
```

# Finding the service at each port

```python
import socket

services = {}

def find_service_name():
    protocolname = 'tcp'
    for port in range(1024):
        try:
            service_name = socket.getservbyport(port, protocolname)
            services[port] = service_name
        except OSError:
            # Skipping ports that do not have an associated service name
            continue

    # Print the results after collecting all service names
    for port, service_name in services.items():
        print(f"Port: {port} => {service_name}")

find_service_name()
```

# Find port by service name

```python
{service:port for port,service in 
    services.items()}['smtp']
```

# Requests Library

```python
import requests
import json

def get_request():
    response = requests.get('https://jsonplaceholder.typicode.com/todos')
    if response.status_code == 200:
        print("Response from server:")
        print(response.json())
        
        users = response.json()
        for user in users:
            print(user['id'])
            print('====')
            print(user['title'])        
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")

get_request()
```

# Web Scraping using BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

def web_scraping():
    url = 'https://www.facebook.com'
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        title = soup.find('title').text
        print(f"Title of the page: {title}")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

if __name__ == "__main__":
    web_scraping()
```

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
    <head><title>The Dormouse's story</title></head>
    <body>
        <p class="title"><b>The Dormouse's story</b></p>
        <p class="story">Once upon a time there were three little sisters; and their names were
            <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
            <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
            <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
            and they lived at the bottom of a well.</p>
        <p class="story">...</p>
    </body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
print("Title of the page:", soup.title.string)
print("First paragraph:", soup.p)
print("All paragraphs:", soup.find_all('p'))
print("All links:", soup.find_all('a'))
print("Link 1:", soup.find(id='link1'))
```
